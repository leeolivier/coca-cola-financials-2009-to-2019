```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read data

third_op_seg = pd.read_excel("/3rd_operation_segment.xlsx")

#convert int to datetime "year" column

third_op_seg["year"] = pd.to_datetime(third_op_seg["year"], format = "%Y")

third_op_seg
```

```python
#check for NAN

third_op_seg.isnull().sum()
```

```python
#describe data

third_op_seg.describe()
```

```python
#group data by "operating segment"

data1 = third_op_seg.groupby("operating segment").sum()
data1
```

```python
#Visualizing the data

data1.plot(figsize = (20, 20))
plt.locator_params(nbins = 20)
plt.title("Coca Cola 3 Months Ended Operating Segments 2009 - 2019", fontsize = 25)
plt.xticks(fontsize = 14, rotation = 90)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 20)
plt.xlabel("Operating Region", fontsize=18)
plt.ylabel("Million $", fontsize = 18);
```

```python
#group data by "year"

df1 = third_op_seg.groupby("year").sum()

#Visualizing the data

df1.plot(figsize = (20, 20))
plt.title("Coca Cola 3 Months Ended Operating Segments 2009 - 2019", fontsize = 25)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 20)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Million $", fontsize = 18);
```

```python
#select rows of interest

select_data1 = third_op_seg[third_op_seg["operating segment"].isin(["Europe Mid East & Africa", "Latin America",
                                                                  "North America", "Pacific"])]
select_data1
```

```python
#pivot table to plot graph of operating segments vs operating income

pivot_select_data1 = select_data1.pivot_table("operating income million $", ["year"], "operating segment")    
    
pivot_select_data1
```

```python
#Visualizing the data

pivot_select_data1.plot(figsize = (20, 20))
plt.title("Coca Cola 3 Months ended Operating Segments Income 2009 - 2019", fontsize = 25)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 20)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Operating Income Million $", fontsize = 18);
```
